```python
# Draw main menu canvas elements
def draw_menu():
    # Set background to black with a size of 1080 x 1920 pixel (whole screen)
    canvas = Canvas(Window, bg="black", width=1920, height=1080)
    canvas.pack()

    # Display the resized background image on the canvas
    canvas.create_image(0, 0, anchor="nw", image=background)

    # Display the resized logo image on the canvas
    canvas.create_image(30, 30, anchor="nw", image=logo)

    # Create a custom font for title
    title_font = ("Helvetica", 24)

    # Display title text in canvas with font 
    canvas.create_text(500, 100, text="Welcome to, LanguageLearned", font=title_font, anchor="center")

    # Create a custom font for smaller text with font Helvetica and size of 18
    lower_font = ("Helvetica", 18)

    # Display lower text in canvas
    canvas.create_text(500, 150, text="Your English language companion!", font=lower_font, anchor="center")
    
    # Function to be called when Button 1 is clicked
    
    # Create Button 1 - conversation
    button1 = tk.Button(Window, text="Conversation - Start a conversation", command=conversation, **button_style)
    button1.pack(pady=10)
    button1.place(relx=0.35, rely=0.4)

    # Create Button 2 - quiz
    button2 = tk.Button(Window, text="Quiz - Test your skills", command=quiz, **button_style)
    button2.pack(pady=10)
    button2.place(relx=0.35, rely=0.5)

    # Create Button 3 - translation
    button3 = tk.Button(Window, text="Translation - Get a translation of a word or sentence", command=translation, **button_style)
    button3.pack(pady=10)
    button3.place(relx=0.35, rely=0.6)

    # Create Button 4 - learn more
    button4 = tk.Button(Window, text="Learn More - Get insights into your given text (English)", command=learnmore, **button_style)
    button4.pack(pady=10)
    button4.place(relx=0.35, rely=0.7)

# Conversation module
def conversation(event=None):
    # Clear the main menu content
    for widget in Window.winfo_children():
        widget.destroy()
    
    # Use model gemini-pro
    model = genai.GenerativeModel('gemini-pro')
    # Start new chat
    chat = model.start_chat()
    
    # Create new cancas
    canvas = Canvas(Window, bg="black", width=1920, height=1080)
    canvas.pack()

    # Display the resized background image on the canvas
    canvas.create_image(0, 0, anchor="nw", image=background)

    # Display the resized logo image on the canvas
    canvas.create_image(30, 30, anchor="nw", image=logo)

    # Create a frame for the chat box
    chat_frame = tk.Frame(canvas, bg="white", width=800, height=800)
    canvas.create_window(500, 400, window=chat_frame, anchor="center")

    # Create a Text widget to display messages
    chat_box = tk.Text(chat_frame, width=50, height=20, wrap=tk.WORD, font=("Helvetica", 12), bg="white", fg="black", 
                       insertbackground="black", borderwidth=2, relief=tk.SOLID, padx=5, pady=5)
    chat_box.pack(padx=10, pady=10)

    # Create a scrollbar for the Text widget
    scrollbar = tk.Scrollbar(chat_frame, command=chat_box.yview)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

    # Configure the Text widget to use the scrollbar
    chat_box.config(yscrollcommand=scrollbar.set)

    # Create an entry widget to input messages
    entry = tk.Entry(chat_frame, width=50)
    entry.pack(padx=10, pady=5)
    
    # Set text widget to be bold
    chat_box.tag_configure("bold", font=("Helvetica", 12, "bold"))
    
    # Display informative message in box 
    chat_box.insert(tk.END, f"Type your first message in the box below to begin conversation...\n")
    
    # When sending a message
    def send_message(event=None):
        # Get the message from the entry box
        message = entry.get()
        
        # If there is a message
        if message:
            # display your message to the chatbox
            chat_box.insert(tk.END, f"You: {message}\n")
            entry.delete(0, tk.END)
            
            response = message
            
            # Give gemini a promt with user message and save response 
            response = chat.send_message("You are currently in conversation mode where you are going to have a casual conversation, do your best to reply like a human would. Keep your message short. When the question is answered give feedback then move to next question. The following message is a message to you: " + response)
            
            # Display bot message in chatbox
            chat_box.insert(tk.END, f"Bot: "+response.text+"\n", "bold")
            
            # Auto-scroll to the end of the chat box for easy use
            chat_box.see(tk.END)
            
    # Bind the <Return> key to the send_message function
    entry.bind("<Return>", send_message)
    
    # Create a button to send messages
    send_button = tk.Button(chat_frame, text="Send", command=send_message, **button_style)
    send_button.pack(side=tk.RIGHT, padx=(10, 5), pady=5)

    # Create a button to send messages
    back_button = tk.Button(chat_frame, text="Back", command=button_back, **button_style)
    back_button.pack(side=tk.LEFT, padx=(5, 10), pady=5)

# Define what back buttons does
def button_back():
    # Clear existing canvas elements
    for widget in Window.winfo_children():
        widget.destroy()
    
    # Draw the main menu
    draw_menu()

# Function to be called when quiz is clicked
def quiz(event=None):
    # Clear existing canvas elements
    for widget in Window.winfo_children():
        widget.destroy()
    
    model = genai.GenerativeModel('gemini-pro')
    chat = model.start_chat()
    canvas = Canvas(Window, bg="black", width=1920, height=1080)
    canvas.pack()

    # Display the resized background image on the canvas
    canvas.create_image(0, 0, anchor="nw", image=background)

    # Display the resized logo image on the canvas
    canvas.create_image(30, 30, anchor="nw", image=logo)

    # Create a frame for the chat box
    chat_frame = tk.Frame(canvas, bg="white", width=800, height=800)
    canvas.create_window(500, 400, window=chat_frame, anchor="center")

    # Create a text widget to display messages
    chat_box = tk.Text(chat_frame, width=50, height=20, wrap=tk.WORD, font=("Helvetica", 12), bg="white", fg="black", 
                       insertbackground="black", borderwidth=2, relief=tk.SOLID, padx=5, pady=5)
    chat_box.pack(padx=10, pady=10)

    # Create a scrollbar for the Text widget
    scrollbar = tk.Scrollbar(chat_frame, command=chat_box.yview)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

    # Configure the Text widget to use the scrollbar
    chat_box.config(yscrollcommand=scrollbar.set)

    # Create an Entry widget to input messages
    entry = tk.Entry(chat_frame, width=50)
    entry.pack(padx=10, pady=5)
    
    chat_box.tag_configure("bold", font=("Helvetica", 12, "bold"))
    
    chat_box.insert(tk.END, f"Type a message like 'ready' to begin!\n")
    
    def send_message(event=None):
        message = entry.get()
        if message:
            chat_box.insert(tk.END, f"You: {message}\n")
            entry.delete(0, tk.END)
            response = message
            response = chat.send_message("be a helpful bot that gives quiz/questions for korean students to learn english (make your message shot).only do one question at a time. Make sure the student is correct or wrong and tell them. The following message is from a korean student: " + response)

            chat_box.insert(tk.END, f"Bot: "+response.text+"\n", "bold")
            
            # Auto-scroll to the end of the chat box
            chat_box.see(tk.END)
            
    # Bind the <Return> event to the send_message function
    entry.bind("<Return>", send_message)
    
    # Create a button to send messages
    send_button = tk.Button(chat_frame, text="Send", command=send_message, **button_style)
    send_button.pack(side=tk.RIGHT, padx=(10, 5), pady=5)  # Align to the left with padding on the left side

    # Create a button to send messages
    back_button = tk.Button(chat_frame, text="Back", command=button_back, **button_style)
    back_button.pack(side=tk.LEFT, padx=(5, 10), pady=5)

# Function to be called when translation is clicked
def translation():
    # Clear existing content
    for widget in Window.winfo_children():
        widget.destroy()
    
    canvas = Canvas(Window, bg="black", width=1920, height=1080)
    canvas.pack()

    # Display the resized background image on the canvas
    canvas.create_image(0, 0, anchor="nw", image=background)

    # Display the resized logo image on the canvas
    canvas.create_image(30, 30, anchor="nw", image=logo)

    # Create a frame for the chat box
    chat_frame = tk.Frame(canvas, bg="white", width=800, height=800)
    canvas.create_window(500, 400, window=chat_frame, anchor="center")

    # Create a Text widget to display messages
    chat_box = tk.Text(chat_frame, width=50, height=20, wrap=tk.WORD, font=("Helvetica", 12), bg="white", fg="black", 
                       insertbackground="black", borderwidth=2, relief=tk.SOLID, padx=5, pady=5)
    chat_box.pack(padx=10, pady=10)

    # Create a scrollbar for the Text widget
    scrollbar = tk.Scrollbar(chat_frame, command=chat_box.yview)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

    # Configure the Text widget to use the scrollbar
    chat_box.config(yscrollcommand=scrollbar.set)

    # Create an Entry widget to input messages
    entry = tk.Entry(chat_frame, width=50)
    entry.pack(padx=10, pady=5)
    
    chat_box.tag_configure("bold", font=("Helvetica", 12, "bold"))
    
    chat_box.insert(tk.END, f"Type a message to translate...\n")
    
    def send_message(event=None):
        message = entry.get()
        if message:
            chat_box.insert(tk.END, f"You: {message}\n")
            entry.delete(0, tk.END)
            
            # If message contains english letters give back a translation to korean otherwise it would be in korean so give back english
            if has_english_letter(message):
                response = GoogleTranslator(source='auto',target='ko').translate(message)
            else:
                response = GoogleTranslator(source='auto',target='en').translate(message)

            chat_box.insert(tk.END, f"Bot Translation: "+response+"\n", "bold")
            
            # Auto-scroll to the end of the chat box
            chat_box.see(tk.END)
            
    # Bind the <Return> event to the send_message function
    entry.bind("<Return>", send_message)
    
    # Create a button to send messages
    send_button = tk.Button(chat_frame, text="Send", command=send_message, **button_style)
    send_button.pack(side=tk.RIGHT, padx=(10, 5), pady=5)  # Align to the left with padding on the left side

    # Create a button to send messages
    back_button = tk.Button(chat_frame, text="Back", command=button_back, **button_style)
    back_button.pack(side=tk.LEFT, padx=(5, 10), pady=5)

# Function to be called when learnmore is clicked
def learnmore():
    # Clear existing content
    for widget in Window.winfo_children():
        widget.destroy()
    
    canvas = Canvas(Window, bg="black", width=1920, height=1080)
    canvas.pack()

    # Display the resized background image on the canvas
    canvas.create_image(0, 0, anchor="nw", image=background)

    # Display the resized logo image on the canvas
    canvas.create_image(30, 30, anchor="nw", image=logo)

    # Create a frame for the chat box
    chat_frame = tk.Frame(canvas, bg="white", width=800, height=800)
    canvas.create_window(500, 400, window=chat_frame, anchor="center")

    # Create a Text widget to display messages
    chat_box = tk.Text(chat_frame, width=50, height=20, wrap=tk.WORD, font=("Helvetica", 12), bg="white", fg="black", 
                       insertbackground="black", borderwidth=2, relief=tk.SOLID, padx=5, pady=5)
    chat_box.pack(padx=10, pady=10)

    # Create a scrollbar for the Text widget
    scrollbar = tk.Scrollbar(chat_frame, command=chat_box.yview)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

    # Configure the Text widget to use the scrollbar
    chat_box.config(yscrollcommand=scrollbar.set)

    # Create an Entry widget to input messages
    entry = tk.Entry(chat_frame, width=50)
    entry.pack(padx=10, pady=5)
    
    chat_box.tag_configure("bold", font=("Helvetica", 12, "bold"))
    
    chat_box.insert(tk.END, f"Type a message to gain more insights...\n")
    
    def send_message(event=None):
        message = entry.get()
        if message:
            chat_box.insert(tk.END, f"You: {message}\n")
            entry.delete(0, tk.END)
            
            # Get a cleaned version of user's input
            clean_message = remove_punctuation(message)
            
            # Tokenize the sentence into words
            words = word_tokenize(clean_message)

            # Perform POS tagging
            pos_tags = nltk.pos_tag(words)
            
            doc = nlp(clean_message)

            # Iterate over the entities recognized
            for ent in doc.ents:
                response = f"'{ent.text}' is a, {ent.label_}."

                chat_box.insert(tk.END, f"Bot Insights: "+response+"\n", "bold")
                
            # Print the tokens and their corresponding POS tags
            for token, tag in pos_tags:
                full_tag = tag_mapping.get(tag, tag) 
                
                # Display the word with its tag as a full word for better understanding
                response = f"'{token}' is a, {full_tag}."

                chat_box.insert(tk.END, f"Bot Insights: "+response+"\n", "bold")
            
            # Get sentiment
            predicted_sentiment = analyze_sentiment(clean_message, tfidf_vectorizer, svm_clf)
            
            
            # Display sentiment
            chat_box.insert(tk.END, f"Bot Insights: Predicted sentiment is {predicted_sentiment[0]}\n", "bold")
            
            # Auto-scroll to the end of the chat box
            chat_box.see(tk.END)
            
    # Bind the <Return> key to the send_message function
    entry.bind("<Return>", send_message)
    
    # Create a button to send messages
    send_button = tk.Button(chat_frame, text="Send", command=send_message, **button_style)
    send_button.pack(side=tk.RIGHT, padx=(10, 5), pady=5)  # Align to the left with padding on the left side

    # Create a button to send messages
    back_button = tk.Button(chat_frame, text="Back", command=button_back, **button_style)
    back_button.pack(side=tk.LEFT, padx=(5, 10), pady=5)

# return whether the input contains english characters
def has_english_letter(text):
    # Define all the english letters in and caps as well
    english_letters = set('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ')
    
    return any(char in english_letters for char in text)

# Remove puntuation and return
def remove_punctuation(sentence):
    # Create a string containing all punctuation characters
    punctuations = string.punctuation
    
    # Remove punctuation
    return ''.join(char for char in sentence if char not in punctuations)

# Train sentiment analysis model
def train_model():
    # Load the data
    train_data = pd.read_csv('train.csv', encoding='latin-1')
    test_data = pd.read_csv('test.csv', encoding='latin-1')

    # Drop rows with missing values for performance
    train_data.dropna(inplace=True)
    test_data.dropna(inplace=True)

    # Split data into features (X) and labels (y)
    X_train = train_data['text']
    y_train = train_data['sentiment']
    X_test = test_data['text']
    y_test = test_data['sentiment']

    # Vectorize text data using TF-IDF
    tfidf_vectorizer = TfidfVectorizer(max_features=10000)
    X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
    X_test_tfidf = tfidf_vectorizer.transform(X_test)

    # Train a Linear Support Vector Classifier
    svm_clf = LinearSVC()
    svm_clf.fit(X_train_tfidf, y_train)

    # Evaluate the model and print results
    y_pred = svm_clf.predict(X_test_tfidf)
    print(classification_report(y_test, y_pred))
    
    return tfidf_vectorizer, svm_clf

# Get sentiment of user's input
def analyze_sentiment(sentence, tfidf_vectorizer, svm_clf):
    input_sentence = sentence
    
    # Preprocess the input sentence
    preprocessed_sentence = preprocess_text(input_sentence)

    # Vectorize the preprocessed sentence using the same vectorizer used during training
    X_input = tfidf_vectorizer.transform([preprocessed_sentence])

    # Use the trained classifier to predict sentiment
    predicted_sentiment = svm_clf.predict(X_input)
    
    return predicted_sentiment

# Preprocess text and clean
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # Join tokens back into string
    processed_text = ' '.join(filtered_tokens)
    return processed_text

# Create a mapping from abbreviated tags to their full forms for better understanding
tag_mapping = {
    'CC': 'Coordinating conjunction',
    'CD': 'Cardinal number',
    'DT': 'Determiner',
    'EX': 'Existential there',
    'FW': 'Foreign word',
    'IN': 'Preposition or subordinating conjunction',
    'JJ': 'Adjective',
    'JJR': 'Adjective, comparative',
    'JJS': 'Adjective, superlative',
    'LS': 'List item marker',
    'MD': 'Modal',
    'NN': 'Noun, singular or mass',
    'NNS': 'Noun, plural',
    'NNP': 'Proper noun, singular',
    'NNPS': 'Proper noun, plural',
    'PDT': 'Predeterminer',
    'POS': 'Possessive ending',
    'PRP': 'Personal pronoun',
    'PRP$': 'Possessive pronoun',
    'RB': 'Adverb',
    'RBR': 'Adverb, comparative',
    'RBS': 'Adverb, superlative',
    'RP': 'Particle',
    'SYM': 'Symbol',
    'TO': 'to',
    'UH': 'Interjection',
    'VB': 'Verb, base form',
    'VBD': 'Verb, past tense',
    'VBG': 'Verb, gerund or present participle',
    'VBN': 'Verb, past participle',
    'VBP': 'Verb, non-3rd person singular present',
    'VBZ': 'Verb, 3rd person singular present',
    'WDT': 'Wh-determiner',
    'WP': 'Wh-pronoun',
    'WP$': 'Possessive wh-pronoun',
    'WRB': 'Wh-adverb'
}

# Main driver function
if __name__ == '__main__':
    # Import requirments
    from tkinter import *
    from tkinter import ttk
    import tkinter as tk
    from PIL import ImageTk, Image
    import os
    import string
    import pathlib
    import textwrap
    from deep_translator import GoogleTranslator
    import google.generativeai as genai
    import nltk
    from IPython.display import display
    from IPython.display import Markdown
    from nltk.tokenize import word_tokenize
    import spacy
    from textblob import TextBlob
    import pandas as pd
    from sklearn.model_selection import train_test_split
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.svm import LinearSVC
    from sklearn.metrics import classification_report
    import re
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize

    # Config Gemini with key
    genai.configure(api_key='AIzaSyC5sFn19E_honxe7lKxmfGRTzGqVBbvwb0')
    
    # Train model for later use
    tfidf_vectorizer, svm_clf = train_model()
    
    # Load the English language model
    nlp = spacy.load("en_core_web_sm")
    
    # Style for buttons
    button_style = {
        'bg': '#0047A0',        # Background color
        'fg': 'white',       # Text color
        'font': ('Helvetica', 12),  # Font
        'relief': 'ridge',  # Button border style
        'borderwidth': 1,    # Border width     
        'height': 2          # Button height
    }
    
    # Define image locations
    logo1 = 'images/languagelearnedlogo1.png'
    logo2 = 'images/languagelearnedlogo2.png'
    background = 'images/background.png'
    
    # Resize images 
    original_image = Image.open(logo1)
    resized_image = original_image.resize((200, 200))
    
    # Resize images
    original_image22 = Image.open(background)
    background_img = original_image22.resize((1920, 1080))
    
    # Making main window, naming it and setting size
    Window = Tk()
    Window.title("LanguageLearned")
    Window.geometry("800x800")

    # Set the window icon
    Window.iconbitmap("images/logoicon.ico")

    # Convert images to PhotoImage objects
    background = ImageTk.PhotoImage(background_img)
    logo = ImageTk.PhotoImage(resized_image)
    
    # Draw menu
    draw_menu()
    
    # Begin mainloop
    Window.mainloop()
```
